# Few basic operations
Here, you will know about a few basic image operations like resize, filtering, etc.

Resize image

```python
# ___Resize___
image_resized1 = cv2.resize(imagecv, (240, 160))
plt.imshow(cv2.cvtColor(image_resized1, cv2.COLOR_BGR2RGB))
plt.show()

#The image is displayed below
```

```python
image_resized2 = cv2.resize(imagecv, (120, 80))
plt.imshow(cv2.cvtColor(image_resized2, cv2.COLOR_BGR2RGB))
plt.show()
```

Filter

```python
# ___Filter___
#Gaussian Blur
image_filtered = cv2.GaussianBlur(imagecv, (7, 7), 0)
plt.imshow(cv2.cvtColor(image_filtered, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#Save the filtered image
import os
#New image directory
directory = 'drive/My Drive/BSDS300-images'
#Change the current location to the specified directory
os.chdir(directory)
#Save the new image
cv2.imwrite('image_filtered.jpg', image_filtered)
```

### Converting Matrix to Image

Useful in implementing transformation/operation on your own.

```python
# Converting Matrix to Image


# Example grayscale matrix (replace this with your own data)
grayscale_matrix = np.array([
    [0, 128, 255],
    [64, 192, 32],
    [96, 16, 144]
], dtype=np.uint8)

# Example RGB matrix (replace this with your own data)
rgb_matrix = np.array([
    [(255, 0, 0), (0, 255, 0), (0, 0, 255)],
    [(128, 128, 0), (0, 128, 128), (128, 0, 128)],
    [(255, 255, 255), (0, 0, 0), (128, 128, 128)]
], dtype=np.uint8)

# Display grayscale image using Matplotlib
plt.imshow(grayscale_matrix, cmap='gray')
plt.title("Gray Scale Image")
plt.axis('off')  # Turn off axes
plt.show()

# Display RGB image using Matplotlib
plt.imshow(rgb_matrix)
plt.title("RGB Image")
plt.axis('off')  # Turn off axes
plt.show()
```

### Transforming an existing image using matrix operation

```python

# imagepath = 'drive/My Drive/BSDS300-images/images/12003.jpg'
# imagepil = Image.open(imagepath)

# Convert PIL image to a NumPy array
image_array = np.array(imagepil)

# Image as matrix representation
print(image_array)

# Adding Constant Value to each value
constant_value = 50
added_image_array = image_array + constant_value

# Image as matrix representation
print(added_image_array)

# Ensure the pixel values stay within the valid range (0-255)
added_image_array = np.clip(added_image_array, 0, 255)

# Convert the NumPy array back to a PIL image
added_pil_image = Image.fromarray(added_image_array.astype(np.uint8))

# Display the original and added images
# imagepil.show(title='Original Image')
# added_pil_image.show(title='Added Image')
plt.imshow(added_pil_image)
```
