```python
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def image_contrast(rgb_image):
    rgb_image=np.asarray(rgb_image)
    hsv_image_array = cv2.cvtColor(rgb_image,cv2.COLOR_BGR2HSV)
    value = hsv_image_array[:, :, 2]
    
    # Apply histogram equalization to the Value component
    enhanced_value = cv2.equalizeHist(value)

    # Update the Value component with the enhanced values
    hsv_image_array[:, :, 2] = enhanced_value

    # Convert the enhanced HSV image back to BGR
    enhanced_bgr_image = cv2.cvtColor(hsv_image_array, cv2.COLOR_HSV2BGR)
    plt.subplot(121),plt.imshow(rgb_image)
    plt.title('Input Image'), plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(enhanced_bgr_image)
    plt.title('Output Image'), plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
imagepath1 = '/Users/vaibhavrathore1999/Downloads/MS/EE 610 Image Processing/Assignments/comp_assgn2_images/low-contrast-photography-1.jpg'
imagecv = cv2.imread(imagepath1) #Importing Image
image_contrast(imagecv)

imagepath2 = '/Users/vaibhavrathore1999/Downloads/MS/EE 610 Image Processing/Assignments/comp_assgn2_images/nature.jpg'
image_contrast(cv2.imread(imagepath2))
```

```python
original_image = cv2.imread(imagepath1)

# Convert to HSV color space
hsv_image_array = cv2.cvtColor(original_image, cv2.COLOR_BGR2HSV)

# Extract the Value channel
value = hsv_image_array[:, :, 2]
# Perform contrast stretching on the Value channel
min_value = np.min(value)
max_value = np.max(value)
stretched_value_channel = ((value - min_value) / (max_value - min_value)) * 255

# Update the Value channel with the stretched values
hsv_image_array[:, :, 2] = stretched_value_channel.astype(np.uint8)

# Convert back to RGB color space
contrast_stretched_image = cv2.cvtColor(hsv_image_array, cv2.COLOR_HSV2BGR)

plt.subplot(121),plt.imshow(contrast_stretched_image)
plt.title('Contrast Stretched Image'), plt.xticks([]), plt.yticks([])

```
